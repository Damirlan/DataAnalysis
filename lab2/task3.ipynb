# Задача 3: Определение некорректных записей посещаемости

Нужно посчитать, сколько значений в наборе являются выбросами по правилу:

- находим квартиль Q1, медиану Q2 и квартиль Q3;
- считаем IQR = Q3 − Q1;
- выбросами считаются значения меньше Q1 − 1.5·IQR или больше Q3 + 1.5·IQR.

Далее шаги решения в отдельных ячейках.


**Шаг 1. Считывание входных данных**

Формат: в первой строке N, далее N чисел. Для демонстрации используем пример из условия.


```python
# Пример входных данных
input_text = "9\n25\n24\n19\n99\n18\n17\n0\n21\n26\n"

# Парсинг данных
lines = [line.strip() for line in input_text.strip().splitlines() if line.strip() != ""]
N = int(lines[0])
values = list(map(int, lines[1:1 + N]))

print("N =", N)
print("values =", values)
```

**Шаг 2. Сортировка и вычисление медианы (Q2)**

Сначала сортируем данные и вычисляем медиану упорядоченного набора.


```python
def median(sorted_list):
    n = len(sorted_list)
    if n % 2 == 1:
        return sorted_list[n // 2]
    return (sorted_list[n // 2 - 1] + sorted_list[n // 2]) / 2

sorted_values = sorted(values)
Q2 = median(sorted_values)

print("Отсортированные данные:", sorted_values)
print("Q2 (медиана) =", Q2)
```

**Шаг 3. Вычисление Q1 и Q3**

Разделяем упорядоченный набор на нижнюю и верхнюю половины
(если N нечётное, центральный элемент не входит ни в одну половину),
затем берём медианы этих половин.


```python
n = len(sorted_values)
if n % 2 == 1:
    lower_half = sorted_values[:n // 2]
    upper_half = sorted_values[n // 2 + 1:]
else:
    lower_half = sorted_values[:n // 2]
    upper_half = sorted_values[n // 2:]

Q1 = median(lower_half)
Q3 = median(upper_half)

print("Нижняя половина:", lower_half)
print("Верхняя половина:", upper_half)
print("Q1 =", Q1)
print("Q3 =", Q3)
```

**Шаг 4. Расчёт IQR и границ выбросов**

IQR = Q3 − Q1. Далее вычисляем нижнюю и верхнюю границы.


```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("IQR =", IQR)
print("Нижняя граница =", lower_bound)
print("Верхняя граница =", upper_bound)
```

**Шаг 5. Подсчёт некорректных значений**

Считаем все значения, которые выходят за пределы допустимого диапазона.


```python
outliers = [v for v in values if v < lower_bound or v > upper_bound]
count_outliers = len(outliers)

print("Некорректные значения:", outliers)
print("Ответ:", count_outliers)
```
