```python
# Ячейка 1: Определение функции сжатия
def compress_string(s):
    if len(s) <= 1:
        return s
    
    result = []
    count = 1
    
    for i in range(1, len(s)):
        if s[i] != s[i-1]:
            if count > 1:
                result.append(str(count))
            result.append(s[i-1])
            count = 1
        else:
            count += 1
            
    if count > 1:
        result.append(str(count))
    result.append(s[-1])
    
    return ''.join(result)
```

```python
# Ячейка 2: Пример ввода и вывода
input_str = 'AAACBBBBBDDADDD'  # Исходная строка
output_str = compress_string(input_str)  # Полученный результат
print(f"Входные данные:\n{input_str}\n")
print(f"Выходные данные:\n{output_str}")
```
