```python
import re

def is_valid_password(password):
    # Проверка минимальной длины
    if len(password) < 12:
        return False
    
    # Регулярные выражения для проверки наличия необходимых символов
    has_uppercase = bool(re.search(r'[A-Z]', password))   # Заглавная буква
    has_lowercase = bool(re.search(r'[a-z]', password))   # Строчная буква
    has_digit = bool(re.search(r'\d', password))          # Цифра
    has_special_char = bool(re.search(r'[!\@\#\$\%\&\*\+]', password))  # Спецсимвол
    
    # Все условия должны выполняться одновременно
    return all([has_uppercase, has_lowercase, has_digit, has_special_char])

# Чтение входных данных
N = int(input("Количество паролей: "))
passwords = [input() for _ in range(N)]

# Проверка всех паролей и вывод результатов
for pwd in passwords:
    print("Valid" if is_valid_password(pwd) else "Invalid")
```
