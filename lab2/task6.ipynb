# Задача 6: Авторы писем в mbox.txt

Нужно:
1) найти адреса всех авторов (строки, начинающиеся с `From `);
2) определить автора, который пишет больше всего писем.

Файл берём по ссылке и работаем с содержимым в памяти.


**Шаг 1. Загрузка данных**

Используем `requests`, как в условии.


```python
# импортируем библиотеку для доступа к файлам в интернете 
import requests 
# в переменной mbox хранится текст для работы 
mbox = requests.get('https://www.py4e.com/code3/mbox.txt').text 
# преобразуем текст в список, где каждый объект — это одна строка в файле 
all_lines = mbox.split('\n') 

print("Всего строк:", len(all_lines))
print("Пример строки:", all_lines[0])
```

**Шаг 2. Извлечение адресов**

Строка автора начинается с `From ` (с пробелом после слова).
Адрес — второе слово в такой строке.


```python
emails = []
for line in all_lines:
    if line.startswith('From '):
        parts = line.split()
        if len(parts) >= 2:
            emails.append(parts[1])

print("Количество найденных писем:", len(emails))
print("Первые 5 адресов:", emails[:5])
```

**Шаг 3. Подсчёт писем по авторам**

Считаем, сколько писем у каждого адреса.


```python
from collections import Counter

email_counts = Counter(emails)
print("Количество уникальных авторов:", len(email_counts))
```

**Шаг 4. Автор с максимальным числом писем**

Находим пару (адрес, количество) с максимальным значением.


```python
top_email, top_count = max(email_counts.items(), key=lambda x: x[1])
print("Автор с максимумом писем:", top_email)
print("Количество писем:", top_count)
```

**Шаг 5. (Опционально) Полный список авторов по убыванию**

Если нужно, можно вывести рейтинг.


```python
sorted_authors = sorted(email_counts.items(), key=lambda x: (-x[1], x[0]))
for email, cnt in sorted_authors[:10]:
    print(email, cnt)
```
