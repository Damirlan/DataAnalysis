```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Formula_One_World_Drivers%27_Champions"
tables = pd.read_html(url)

def is_champions_table(t: pd.DataFrame) -> bool:
    cols = [c[0] if isinstance(c, tuple) else c for c in t.columns]
    cols = [str(c).lower() for c in cols]
    return ('season' in cols) and ('driver' in cols) and ('age' in cols)

champ_tbl = None
for t in tables:
    if is_champions_table(t):
        champ_tbl = t.copy()
        break

if champ_tbl is None:
    raise ValueError("Не удалось автоматически найти таблицу чемпионов. Посмотрите tables и выберите нужную вручную.")

champ_tbl.head()
```
